```python
def extract_features(audio_path, n_mfcc=13, n_chroma=12, n_spectral_contrast=6, target_length=20):
    # Load audio file
    audio, sr = librosa.load(audio_path, sr=None)

    # Extract MFCCs (13 coefficients)
    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)

    # Extract Chroma Features (12 pitch classes)
    chroma = librosa.feature.chroma_stft(y=audio, sr=sr, n_chroma=n_chroma)

    # Extract Spectral Centroid
    spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sr)

    # Extract Spectral Contrast (7 bands)
    spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sr, n_bands=n_spectral_contrast)

    # Extract Spectrogram (Magnitude Spectrogram)
    spectrogram = np.abs(librosa.stft(audio))

    # Truncate or pad features to target_length
    def truncate_or_pad(feature, target_length):
        if feature.shape[1] > target_length:
            return feature[:, :target_length]
        else:
            return np.pad(feature, ((0, 0), (0, target_length - feature.shape[1])), mode='constant')

    mfcc = truncate_or_pad(mfcc, target_length)
    chroma = truncate_or_pad(chroma, target_length)
    spectral_centroid = truncate_or_pad(spectral_centroid, target_length)
    spectral_contrast = truncate_or_pad(spectral_contrast, target_length)
    spectrogram = truncate_or_pad(spectrogram, target_length)

    # Flatten features into a single vector
    feature_vector = np.concatenate([
        mfcc.flatten(),
        chroma.flatten(),
        spectral_centroid.flatten(),
        spectral_contrast.flatten(),
        spectrogram.flatten()
    ])

    return feature_vector
```

```python
def process_audio_directory(input_dir, output_dir):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Process each audio file in the input directory
    for file_name in os.listdir(input_dir):
        if file_name.endswith(('.wav')):  # Add more formats if needed
            file_path = os.path.join(input_dir, file_name)

            # Extract features
            feature_vector = extract_features(file_path)

            # Save feature vector to a file
            output_file_path = os.path.join(output_dir, file_name.replace('.wav', '.npy'))
            np.save(output_file_path, feature_vector)
            print(f"Extracted features and saved: {output_file_path}")
```

```python
input_directory = "/content/drive/MyDrive/SafeReels-664/Processed_Audio"
output_directory = "/content/drive/MyDrive/SafeReels-664/Extracted_Features"
process_audio_directory(input_directory, output_directory)
```

```python
def analyze_directory(directory):

    file_counts = {}
    total_files = 0

    for filename in os.listdir(directory):
        filepath = os.path.join(directory, filename)
        if os.path.isfile(filepath):
            total_files += 1
            _, ext = os.path.splitext(filename)
            ext = ext.lower()  # Normalize to lowercase
            file_counts[ext] = file_counts.get(ext, 0) + 1

    print(f"Total number of files in '{directory}': {total_files}")
    for ext, count in file_counts.items():
        print(f"{ext}: {count}")
```
