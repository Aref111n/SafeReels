```python
fdf['reel_id'] = fdf['reel_id'].astype(str)
fdf['reel_id'] = 'www.facebook.com/reel/' + fdf['reel_id']
```

```python
fdf_filtered = fdf[~fdf['reel_id'].isin(df['reel_id'])]

fdf_filtered
```

```python
fdf_filtered['UnsafeType'].value_counts()
```

```python
df['UnsafeType'].value_counts()
```

```python
suicidal_sample = fdf_filtered[fdf_filtered['UnsafeType'] == 'Suicidal'].sample(n=9, random_state=42)
suicidal_sample
```

```python
adult_sample = fdf_filtered[fdf_filtered['UnsafeType'] == 'Adult'].sample(n=150, random_state=42)
adult_sample
```

```python
harmful_sample = fdf_filtered[fdf_filtered['UnsafeType'] == 'Harmful'].sample(n=150, random_state=42)
harmful_sample
```

```python
final_sample = pd.concat([suicidal_sample, adult_sample, harmful_sample])
final_sample
```

```python
remaining_samples_needed = 336 - len(final_sample)
```

```python
if remaining_samples_needed > 0:
  remaining_sample = fdf_filtered[~fdf_filtered.index.isin(final_sample.index)].sample(n=remaining_samples_needed, random_state=42)
  final_sample = pd.concat([final_sample, remaining_sample])
```

```python
print(final_sample['UnsafeType'].value_counts())
print(len(final_sample))
```

```python
df = pd.concat([df, final_sample])
df
```
